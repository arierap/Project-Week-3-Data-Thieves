### OpenSkills API

*OpenSkills Project* is a public-private partnership lead by the University of Chicago. It’s aim is to improve our understanding of the data labor market skills, normalizing the terms and measuring importance for each data role. We retrieved data about Data Analyst’s related skills through it’s API, which was very straightforward to use. Instructions on how to access this information can be be found on http://dataatwork.org/data/ . 

### Linkedin Webscrapping

After some brainstorming based on the data extracted from OpenSkills, and looking at some job offers we defined a list of the most relevant skills for a Data Analyst. We then proceeded to scrape Linkedin for all Data Analyst offers from the main tech hubs of Europe. We used the BeautifulSoup Library and for-loops to iterate through every single offer, and regex to count how frequent each skill was mentioned in each job offer.

### Glassdoor Web Scrapping

*Glassdoor* Scrapping came after battling with LinkedIn. Our first intention had been collecting data on the skills real people holding a Data Analyst position had listed on their profiles. When trying to scrape it with BeautifulSoup, it redirected us to the login page, and that was what both frustrated us and inspired us to venture into Selenium.

After playing with it for a bit we managed to get recursively into each profile, but the scrapping of the skills turned out to be  too complicated for our level and time-constraint (xpaths not being consistent across profiles, structure not being standarized, which made loops virtually impossible for our lebel). 

We turned to the LinkedIn API, we registered as a company to be able to access it and profit from the profile oriented features, however we were only granted access to the basic profile information, which did not include skills. 
Defeated and tired we had the idea to find information about salaries on Glassdoor.

After what we had learnt trying to scrape LinkedIn with Selenium it proved quite easy, In two hours and after a lot of trial and error, we managed to log in, navigate the several result pages and collect data for job position, company and salary for the 7 cities we were going to base our study on. We fit it all into a Pandas DataFrame and got ready for cleaning. 


### Data Wrangling

Once we compiled all the data, we thought that our next step for the project was going to be smooth, but data cleaning turned to have more complexity that we anticipated. 

Firstly, we had assign a temporality for each job position, since some salaries were posted as yearly, monthly or even hourly. Secondly, we calculated the mean for each salary because in some cases the data for salaries appeared in ranges.

Finally, we calculated the annual salary for each job position based on temporality and the mean salary.
After cleaning both datasets, we aimed to synthesize the information contained in the job title in a column that would indicate the level of experience from Intern to Senior based on keywords found on the job title.

The main problem was that for the intermediate role between Junior and Senior, we could not assume that any title not containing those words was for an intermediate position, so we labelled it as Not Determined.
