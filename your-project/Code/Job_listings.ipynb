```python
import pandas as pd
```

```python
joblisting_cities = pd.read_csv("../Data/raw_dataframes/joblisting_cities.csv")
```

```python
new_column = []
for index,i in enumerate (joblisting_cities['job_offer']):
    string = joblisting_cities['job_offer'][index]
    str_list = string.split(' ')
    if 'Senior' in str_list:
        new_column.append('Senior Position')
    elif 'Junior' in str_list:
        new_column.append('Junior Position')
    elif 'Intern' in str_list:
        new_column.append('Internship')
    elif 'Becario' in str_list:
        new_column.append('Internship')
    elif 'Internship' in str_list:
        new_column.append('Internship')
    else:
        new_column.append('Not determined')
new_column
```

```python
joblisting_cities["Seniority"] = pd.DataFrame(new_column)
```

```python
joblisting_cities[joblisting_cities["Seniority"]=="Internship"].sort_values(by="xp",ascending=False).head(30)
```

```python
joblisting_cities.to_csv("joblisting_final.csv")
```
