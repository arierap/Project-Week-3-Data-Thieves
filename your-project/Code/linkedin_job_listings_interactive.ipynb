```python
import re
import time
```

```python
info = {'link':[],"company":[],"redshift":[],"sql":[],"kpi":[],"py":[],"xp":[],"R":[],"business":[],"insights":[],"tableau":[]
       ,"looker":[],"qlik":[],"pbi":[],"ab":[],"stakeholders":[],"quant":[],"oper":[],"mgmt":[]}
df_2 = pd.DataFrame(data=info)
counter = 0
for i in links_list:
    info["link"].append(i)
    offer_url = f"{i}"
    response = requests.get(offer_url)
    print(response)
    counter += 1
    print(f"Scraping offer: {counter}/{pages*25}")
    #time.sleep(1)
    offer = BeautifulSoup(response.content, "html.parser")
    offer_2 = offer.find_all("div", class_=re.compile("description"))
    offer_2 = offer_2[0].prettify()
    try:
        company = offer.select("h3 a")[0].get_text()
        info["company"].append(company)
    except:
        company = "Error scraping company"
        info["company"].append(company)
    redshift = len(re.findall("[Aa]mazon [Rr]edshift",offer_2))
    info["redshift"].append(redshift)
    sql = len(re.findall("[Ss][Qq][Ll]",offer_2))
    info["sql"].append(sql)
    kpi = len(re.findall("[Kk][Pp][Ii]",offer_2))
    info["kpi"].append(kpi)
    xp = len(re.findall("[Ee]xperience",offer_2))
    info["xp"].append(xp)
    py = len(re.findall("[P]ython",offer_2))
    info["py"].append(py)
    R = len(re.findall(r"\b[R]\b",offer_2))
    info["R"].append(R)
    business = len(re.findall("[Bb]usiness",offer_2))
    info["business"].append(business)
    insights = len(re.findall("[Ii]nsight",offer_2))
    info["insights"].append(insights)
    tableau = len(re.findall("[Tt]ableau",offer_2))
    info["tableau"].append(tableau)
    looker = len(re.findall("[Ll]ooker",offer_2))
    info["looker"].append(looker)
    qlik = len(re.findall("[Qq]lik",offer_2))
    info["qlik"].append(qlik)
    pbi = len(re.findall("[Pp]ower[Bb][Ii]",offer_2))
    info["pbi"].append(pbi)
    ab = len(re.findall("[Aa]/[Bb]",offer_2))
    info["ab"].append(ab)
    stakeholders = len(re.findall("[Ss]takeholder",offer_2))
    info["stakeholders"].append(stakeholders)
    quant = len(re.findall("[Qq]uantitative",offer_2))
    info["quant"].append(quant)
    oper = len(re.findall("[Oo]peration",offer_2))
    info["oper"].append(oper)
    mgmt = len(re.findall("[Mm]anag",offer_2))
    info["mgmt"].append(mgmt)
```

```python
df_2 = pd.DataFrame(data=info)
df_2
```

```python
df_final = df.merge(df_2,how="left")
```

```python
df_final["location"]= f"London"
df_final.to_csv(f"London_offers.csv")
```
